# Exploratory Data Analysis (EDA)

This notebook explores the synthetic customer feedback dataset.  We load the raw CSV file, examine basic statistics, check for duplicates and empty rows, and visualise class counts and message length distributions.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/raw/customer_feedback.csv')
# Display first few rows
df.head()
```

```python
# Dataset shape
print(f"Shape: {df.shape}")

# Class counts
class_counts = df['category'].value_counts()
print(class_counts)

# Plot class distribution
class_counts.plot(kind='bar', color='skyblue', title='Class distribution of synthetic customer feedback')
plt.xlabel('Category')
plt.ylabel('Number of samples')
plt.tight_layout()
plt.savefig('../reports/figures/class_distribution.png')
plt.show()
```

```python
# Message length distribution
message_lengths = df['text'].str.len()

plt.figure()
plt.hist(message_lengths, bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of message lengths (characters)')
plt.xlabel('Length of message')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('../reports/figures/message_length_distribution.png')
plt.show()
```

```python
# Check for duplicates and empty texts
duplicates = df.duplicated().sum()
empty_texts = (df['text'].str.strip() == '').sum()
print(f"Number of duplicate rows: {duplicates}")
print(f"Number of empty text rows: {empty_texts}")
```
